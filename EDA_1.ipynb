```python
import pandas as pd 
```

```python
df_1=pd.read_csv(r"C:\Users\DELL\Documents\GitHub\imdb_project\Datafinal_tmdb_data_2000.csv.gz")
df_1.head()
```

```python
df_2=pd.read_csv(r"C:\Users\DELL\Documents\GitHub\imdb_project\Datafinal_tmdb_data_2001.csv.gz")
df_2.head()
```

```python
# Merge the dataframes based on the common column
df = pd.concat([df_1, df_2])
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python

#movies had at least some valid financial information
valid_movies = df[(df['budget'] > 0) & (df['revenue'] > 0)]
invalid_movies = df[(df['budget'] <= 0) | (df['revenue'] <= 0)]

num_valid_movies = len(valid_movies)
num_invalid_movies = len(invalid_movies)

plt.bar(['Valid Movies', 'Invalid Movies'], [num_valid_movies, num_invalid_movies])
plt.xlabel('Movies')
plt.ylabel('Number of Movies')
plt.title('Movies with Valid and Invalid Financial Information')
plt.show()
```

```python
df.info()
```

```python
#Please exclude any movies with 0's for budget AND revenue
df = df[(df['budget'] > 0) & (df['revenue'] > 0)]
df.info()
```

```python
#How many movies are there in each of the certification categories (G/PG/PG-13/R)?
certification_counts = df['certification'].value_counts()
plt.bar(certification_counts.index, certification_counts.values)
plt.xlabel('Certification')
plt.ylabel('Number of Movies')
plt.title('Number of Movies in Each Certification Category')
plt.show()
```

```python
#What is the average revenue per certification category?
average_revenue = df.groupby('certification')['revenue'].mean()
average_revenue.plot(kind='bar')
plt.xlabel('Certification Category')
plt.ylabel('Average Revenue')
plt.title('Average Revenue per Certification Category')
plt.show()
```

```python
#What is the average budget per certification category?
average_budget = df.groupby('certification')['budget'].mean()
average_budget.plot(kind='bar')
plt.xlabel('Certification Category')
plt.ylabel('Average budget')
plt.title('Average budget per Certification Category')
plt.show()
```

```python
#saving the editted data frame 
df.to_csv(r"C:\Users\DELL\Documents\GitHub\imdb_project\Data\tmdb_results_combined.csv.gz",compression='gzip',index = False)
```
