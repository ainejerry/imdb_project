```python
import pandas as pd
```

```python
#loading the title basics data frame into pandas 
df_basics = pd.read_csv(r"C:\Users\DELL\Documents\GitHub\imdb_project\Data\basics_data.csv.gz", low_memory = False)
df_basics.head()
```

```python
filtered_basics=df_basics[['tconst','primaryTitle','startYear','runtimeMinutes']]
filtered_basics.head()
```

```python
## create a col with a list of genres
df_basics['genres_split'] = df_basics['genres'].str.split(',')
df_basics.head()
```

```python
exploded_genres = df_basics.explode('genres_split')
exploded_genres
```

```python
unique_genres = sorted(exploded_genres['genres_split'].unique())
print(unique_genres)
```

```python
#genres dataframe
title_genres=exploded_genres[['tconst','genres_split']].copy()
title_genres.head()
```

```python
## Making the genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
## make new integer genre_id and drop string genres
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres= title_genres.drop(columns='genres_split')
```

```python
title_genres.head()
```

```python
genres=pd.DataFrame({'Genre_Name':genre_map.keys(),
                     'Genre_ID':genre_map.values()})
genres.head()
```

```python
tmdb_df=pd.read_csv(r"C:\Users\DELL\Documents\GitHub\imdb_project\Data\tmdb_results_combined.csv.gz")
tmdb_df.head()
```

```python
#to have matching column names
tmdb_df.rename(columns={'imdb_id':'tconst'},inplace=True)
tmdb_df.head()
```

```python
#filtered tmdb
filtered_tmdb=tmdb_df[['tconst','revenue','budget','certification']]
filtered_tmdb.head()
```

```python
import pymysql
pymysql.install_as_MySQLdb()
from sqlalchemy import create_engine
from sqlalchemy_utils import create_database, database_exists
from sqlalchemy.types import *
```

```python
#connection string and engine
connection = "mysql+pymysql://root:1965@localhost/movies"
engine = create_engine(connection)
```

```python
if database_exists(connection) == False:
  create_database(connection)
else:
  print('The database already exists')
```

```python
## Calculate max string lengths for object columns
key_len = filtered_basics['tconst'].fillna('').map(len).max()
title_len = filtered_basics['primaryTitle'].fillna('').map(len).max()
## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": VARCHAR(255), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'endYear':Float(),
    'runtimeMinutes':Integer()}
```

```python
# Save to sql with dtype and index=False
filtered_basics.to_sql('title_basics',engine,dtype=df_schema,if_exists='replace',index=False)
                       
```

```python
q = """SHOW TABLES;"""
pd.read_sql(q, engine)

```

```python
## Calculate max string lengths for object columns in the rates table
key_len = rates['tconst'].fillna('').map(len).max()

## Create a schema dictonary using Sqlalchemy datatype objects
df_schema = {
    "tconst": String(key_len+1), 
    'averageRating':Float(),
    'numVotes':Integer()}
```
